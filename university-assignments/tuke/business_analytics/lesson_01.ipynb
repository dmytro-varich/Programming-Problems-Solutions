```python
# import librires 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Vytvorte boxplot-y červenej farby, zobrazujúci hodnoty numerického atribútu time_left
# pre rôzne podmnožiny rozdelené podľa hodnôt atribútu answer.

sns.boxplot(x='answer', y='time_left', data=dfsolutions, color='red')
plt.show()
```

```python
# Vytvorte histogram zelenej farby, ktorý zobrazí hodnoty numerického atribútu
# time_left len záznamov, kde subject_id = 23. Do grafu pridajte tiež hrubú, prerušovanú
# horizontálnu čiaru na hodnote 5.
time_left_filtered = dfsolutions[dfsolutions['subject_id'] == 23]['time_left']
plt.hist(time_left_filtered, color='green', edgecolor='black')
plt.axhline(y=5, color='red', linestyle='--', linewidth=2)
plt.xlabel("Time Left")
plt.ylabel("Frequency")
plt.show()
```

```python
# Nájdite spôsob vykreslenia viacerých grafov vedľa seba, ktoré budú znázorňovať atribút
# start v závislosti od atribútu stop, pričom pôjde o podgrafy podľa hodnôt atribútu answer. 

# Unique values of the ANSWER attribute
unique_answers = dfsolutions['answer'].unique()

# Create a canvas with subtraffers based on the number of unique values in Answer
fig, ax = plt.subplots(1, len(unique_answers), figsize=(15, 5), sharey=True)

# We pass for each unique value of ANSWER and build the corresponding schedule
for i, answer in enumerate(unique_answers):
    subset = dfsolutions[dfsolutions['answer'] == answer]  # Subraking by Asswer
    ax[i].plot(subset['start'], subset['stop'])
    ax[i].set_title(f'Answer: {answer}')
    ax[i].set_xlabel('Start')
    
ax[0].set_ylabel('Stop')  # General label for axis y

plt.tight_layout() # Automatically adjusts indentation
plt.show()
```

```python
# Príklad na funkciu, cyklus a podmienku:
# Vytvorte funkciu s názvom funkcia, ktorá bude obsahovať vstupné premenné c,d,n.
# Vytvorte vektor f, ktorý bude nadobúdať hodnoty od 1 po n s krokom 0.5. Použitím
# cyklu prejdite všetky čísla vektora f a pomocou vetvenia urobte nasledovne operácie:
# • Ak je aktuálne číslo totožné s c, nahraďte ho na výstupe znakom „A“
# • Ak je aktuálne číslo totožné s d, nahraďte na výstupe znakom „B“
# • Inak vypíšte normálnu hodnotu aktuálneho čísla z vektora f
# Na záver spustite vytvorenú funkciu s vami (vhodne) zvolenými vstupnými
# premennými.
from numpy import arange
from decimal import Decimal

def funkcia(c: int | float, d: int | float, n: int | float) -> list: 
    f = list(arange(1, n, 0.5))
    for index, x in enumerate(f): 
        if Decimal(x) == Decimal(c): f[index] = "A"
        elif Decimal(x) == Decimal(d): f[index] = "B"
    return f

result = funkcia(1, 3, 5)
print(result)
```

```python
quakes = pd.read_csv("https://peter.butka.website.tuke.sk/res/quakes.csv")
quakes.head()
```

```python
# Prvý graf (horný) bude histogram atribútu mag s nastavením 15 hraničných hodnôt,
# pričom bude modrej farby, ako aj s popisom osi x "Magnitude of quakes", osi y
# "Number of quakes", hlavného nadpisu "Histogram of quake magnitudes". Pridajte ku
# grafu vertikálnu prerušovanú čiaru pre hodnotu magnitúdy 4.8 červenej farby a hrúbky 3.

plt.hist(quakes['mag'], color='blue', edgecolor='black', bins=15)
plt.axvline(x=4.8, color='red', linestyle='--', linewidth=3)
plt.xlabel("Magnitude of quakes")
plt.ylabel("Number of quakes")
plt.title("Histogram of quake magnitudes")
plt.show()
```

```python
# Druhý graf (stredný) bude boxplot atribútu depth zelenej farby s popisom osi y "Depth"
# a hlavným nadpisom "Depth of quakes". Pridajte ku grafu čiaru hrúbky 5 žltej farby pre
# hodnotu atribútu depth rovnú 400. 

sns.boxplot(x='depth', data=quakes, color='green')
plt.axvline(x=400, color='yellow', linewidth=5)
plt.ylabel("Depth") 
plt.title("Depth of quakes")
plt.show()
```

```python
# Tretí graf (spodný) bude bodový graf pre atribút lat na osi y a long na osi x. Typ bodu
# v grafe bude plný trojuholník a ich farba bude červená. Popisy zodpovedajúcich osí
# budú "Latitude" a "Longitude", hlavný nadpis bude "Positions of quakes". Následne
# vložíte do grafu obdĺžnik (funkcia rect) s hranami modrej farby hrúbky 3 tak, aby
# ohraničoval oblasť s hodnotami lat <-30,-20> a long <172,182>.
from matplotlib.patches import Rectangle

fig, ax = plt.subplots()
plt.scatter(quakes['long'], quakes['lat'], color='r', marker='^')
plt.xlabel("Latitude")
plt.xlabel("Longitude")
plt.title("Positions of quakes")
ax.add_patch(Rectangle((172, -30), 10, 10, lw=3, edgecolor='b', facecolor='none'))
plt.xlim(171, 183) # Set the limits of the X axis
plt.ylim(-31, -19) # Set the limits of the Y axis
plt.grid()
plt.show()
```
