```python
# Vytvorte nový data frame s názvom elnino_2, ktorý bude obsahovať skopírované dáta elnino.
elnino_2 = elnino.copy()
```

```python
# Check, if there are missing values 
print(elnino_2[['zon_winds', 'mer_winds']].isna().sum())
```

```python
"""
Doplňte v elnino_2 chýbajúce hodnoty atribútov podávajúcich informácie o Vetroch
(Zonal Winds, Meridional Winds) tak, že ich nahradíte priemerom (mean) daného
stĺpca. 
"""
elnino_2['zon_winds'] = elnino_2['zon_winds'].fillna(elnino_2['zon_winds'].mean())
elnino_2['mer_winds'] = elnino_2['mer_winds'].fillna(elnino_2['mer_winds'].mean())
```

```python
# Check, if there are missing values 
print(elnino_2[['zon_winds', 'mer_winds']].isna().sum())
```

```python
# Check, if there are missing values
print(elnino_2[['humidity', 'air_temp']].isna().sum())
```

```python
"""
Doplňte v elnino_2 chýbajúce hodnoty atribútov Relatívna vlhkosť (Relative
Humidity) a Teplota vzduchu (Air Temperature) tak, že ich nahradíte hodnotou
mediánu daného stĺpca
"""
elnino_2['humidity'] = elnino_2['humidity'].fillna(elnino_2['humidity'].median())
elnino_2['air_temp'] = elnino_2['air_temp'].fillna(elnino_2['air_temp'].median())
```

```python
# Check, if there are missing values
print(elnino_2[['humidity', 'air_temp']].isna().sum())
```

# Part 4

```python
"""
Vytvorte v dátach elnino_1 nové atribúty s názvami zon_winds_1 a mer_winds_1,
ktoré budú odvodené z atribútov Zon_winds a Mer_winds podľa podmienok
popísaných v informáciách o atribútoch (Attribute Information) na
https://archive.ics.uci.edu/ml/datasets/El+Nino – konkrétne budú obsahovať informáciu
o tom či sa jedná o západný (West, keď je záporná hodnota) alebo východný (East, keď
je kladná hodnota) vietor pre Zon_winds, resp. podobne južný (South, keď je záporná
hodnota) alebo severný vietor (North, keď je kladná hodnota) pre Mer_winds.
"""


elnino_1['zon_winds_1'] = 'west'
elnino_1.loc[elnino_1['zon_winds'] > 0, 'zon_winds_1'] = 'east'  

elnino_1['mer_winds_1'] = 'south'
elnino_1.loc[elnino_1['mer_winds'] > 0, 'mer_winds_1'] = 'north'
```

```python
elnino_1.head()
```

```python
elnino_1.info()
```

```python
# Usporiadajte atribúty dataframu elnino_1 takým spôsobom, že atribút zon_winds sa
# bude nachádzať pri zon_winds_1 a rovnako aj mer_winds pri atribúte mer_winds_1.
new_order =  [col for col in elnino_1.columns if col not in ['zon_winds', 'zon_winds_1', 'mer_winds', 'mer_winds_1']] + \
                ['zon_winds', 'zon_winds_1', 'mer_winds', 'mer_winds_1'] 

elnino_1 = elnino_1[new_order]
```

```python
elnino_1.head()
```

```python
# Opäť vypíšte informácie o dátach elnino_1 a upravte podľa potreby dátový typ novo
# vytvorených atribútov na kategoriálny atribút (typ factor v R, typ category v Pythone).
elnino_1['zon_winds_1'] = elnino_1['zon_winds_1'].astype('category')
elnino_1['mer_winds_1'] = elnino_1['mer_winds_1'].astype('category')
```

```python
elnino_1.info()
```

# Part 5

```python
"""
Normalizujte atribút Relatívna vlhkosť (Humidity) postupom tzv. štandardizácie
a atribút Teplota vzduchu (Air Temperature) na rozsah <-1,1> pomocou min-max
normalizácie (Pomôcka: napr. pre R existuje pre štandardizáciu scale funkcia priamo
v base inštalácii (ak nastavíme center a scale parametre na TRUE), v balíku clusterSim
funkcia data.Normalization poskytuje štandardizáciu pre nastavenie type=„n1“, viď.
https://cran.r-project.org/web/packages/clusterSim/clusterSim.pdf, či pre min-max
škálovanie existuje v balíčku scales funkcia rescale (stačí nastaviť správne do akého
rozsahu cez to parameter); v Pythone existujú vhodné funkcie v sklearn.preprocessing
ako napríklad scale a minmax_scale, prípadne sami implementujte vhodný vzorec
výpočtu, alebo použite iné dostupné knižnice).
"""

scaler_standard = StandardScaler()
elnino_1['humidity_standardized'] = scaler_standard.fit_transform(elnino_1[['humidity']])

scaler_minimax = MinMaxScaler(feature_range=(-1, 1))
elnino_1['air_temp_normalized'] = scaler_minimax.fit_transform(elnino_1[['air_temp']])

print(elnino_1[['humidity', 'humidity_standardized', 'air_temp', 'air_temp_normalized']])
```

```python
"""
Vytvorte nový atribút v dátach elnino_1 s názvom s_s_temp_1, kde budú hodnoty
(s_s_temp) rozdelené do 4 rovnako veľkých intervalov (Ekvidištančná diskretizácia).
Vypíšte taktiež hranice intervalov spolu s početnosťou záznamov, ktoré do nich
spadajú.
"""

elnino_1['s_s_temp_1'], bins = pd.cut(elnino_1['ss_temp'], bins=4, retbins=True) 

print("hranice intervalov:", bins)

value_counts = elnino_1['s_s_temp_1'].value_counts().sort_index()
print(value_counts)
```

```python
elnino_1[['ss_temp', 's_s_temp_1']].head()
```

```python
"""
Vytvorte nový atribút v dátach elnino_1 s názvom s_s_temp_2, kde budú hodnoty
(s_s_temp) rozdelené do 3 intervalov vyjadrených triedou od 1,2,3. Hranice intervalov
zvoľte 20,24,28 a 32. Vypíšte taktiež početnosti záznamov, ktoré spadajú do tried 1,2,3. 
"""
bins = [20, 24, 28, 32]
elnino_1['s_s_temp_2'] = pd.cut(elnino_1['ss_temp'], bins=bins, labels=['1', '2', '3']) 

value_counts = elnino_1['s_s_temp_2'].value_counts().sort_index()
print(value_counts)
```

```python
elnino_1[['ss_temp', 's_s_temp_2']].head()
```

```python
"""
Vytvorte nový atribút v dátach elnino_1 s názvom s_s_temp_3, kde budú hodnoty
(s_s_temp) pridelené do 5 intervalov s rôznou šírkou, ale rovnakou početnosťou
záznamov (Ekvifrekvenčná diskretizácia). Vypíšte taktiež početnosti záznamov, ktoré
spadajú do vytvorených intervalov. 
"""
elnino_1['s_s_temp_3'] = pd.qcut(elnino_1['ss_temp'], q=5)

value_counts = elnino_1['s_s_temp_3'].value_counts().sort_index()
print(value_counts)
```
