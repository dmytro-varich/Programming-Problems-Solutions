```python
from currency_converter import CurrencyConverter

c = CurrencyConverter()

for city, table in csv_files.items():
    if city in CZECH_CITY: 
        table['cena'] = table['cena'].apply(lambda x: c.convert(x, 'CZK', 'EUR') if pd.notnull(x) else None)  
        
    table['mesto'] = city.capitalize()
    table['krajna'] = 'Czech Republic' if city in CZECH_CITY else 'Slovakia'

print(csv_files)
```

```python
predaje = pd.concat([table for table in csv_files.values()], axis=0)
```

```python
predaje['mesiac'] = pd.to_datetime(predaje['cas_nakupu']).dt.strftime("%B")
predaje['štvrťrok'] = pd.to_datetime(predaje['cas_nakupu']).dt.to_period('Q')
```
