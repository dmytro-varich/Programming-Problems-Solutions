# Autonomous work 1
Load the `paris.jpg` image from example files in Teams and apply histogram equalization to R, G and B bands separately. Reassemble the adjusted image and visualize.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from skimage import io, exposure
from cv2 import merge

image = io.imread('/content/paris.jpg')

fig = plt.figure(figsize=(15, 10))
axes = fig.subplots(2, 3)

axes[0, 0] = plt.subplot(2, 3, 1)
axes[0, 1] = plt.subplot(2, 3, 2)
axes[0, 2] = plt.subplot(2, 3, 3)
axes[1, 0] = plt.subplot(2, 3, 4)
axes[1, 1] = plt.subplot(2, 3, 5)
axes[1, 2] = plt.subplot(2, 3, 6)

image_eq_R = exposure.equalize_hist(image[:,:,0])
axes[0, 0].imshow(image_eq_R, cmap='gray')
axes[0, 0].set_title('Equalized image adjust for Red Color')

image_eq_G = exposure.equalize_hist(image[:,:,1])
axes[0, 1].imshow(image_eq_G, cmap='gray')
axes[0, 1].set_title('Equalized image adjust for Green Color')

image_eq_B = exposure.equalize_hist(image[:,:,2])
axes[0, 2].imshow(image_eq_B, cmap='gray')
axes[0, 2].set_title('Equalized image adjust for Blue Color')

axes[1, 0].hist(image_eq_R.ravel(), bins=20, histtype='bar', color='red')
axes[1, 0].set_title('Equalized histogram adjust for Red Color')

axes[1, 1].hist(image_eq_G.ravel(), bins=20, histtype='bar', color='green')
axes[1, 1].set_title('Equalized histogram adjust for Green Color')

axes[1, 2].hist(image_eq_B.ravel(), bins=20, histtype='bar', color='blue')
axes[1, 2].set_title('Equalized histogram adjust for Blue Color')

plt.figure(figsize=(15, 10))
plt.subplot(121)
plt.title('Original image')
plt.imshow(image)
plt.subplot(122)
new_image = merge([image_eq_R, image_eq_G, image_eq_B])
plt.title('Equalized image')
plt.imshow(new_image)
```

# Autonomous work 2
Write your own modification of the brightness curve that will produce image negative.

```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.data as data

plt.figure(figsize=(20, 15))

plt.subplot(221)
plt.title('Original Moon')
plt.imshow(data.moon(), cmap='gray')

plt.subplot(222)
plt.title('Negative Moon')

# Load the moon image and scale it to the range [0, 1]
im = data.moon() / 255

# Apply the custom image negative function to the image
adj_im = 1 - im

# Display the inverse moon image
plt.imshow(adj_im, cmap='gray')

# Subplot 3: Create the histogram of the original moon image
plt.subplot(223)
plt.hist(data.moon().ravel(), bins=20, histtype='bar', color='red')
plt.title("Original Histogram")
plt.xlabel("Brightness Levels")
plt.ylabel("Number of Pixels")

# Subplot 4: Create the histogram of the inverse moon image
plt.subplot(224)
plt.hist((adj_im * 255).ravel(), bins=20, histtype='bar', color='red')
plt.title("Inverse Histogram")
plt.xlabel("Brightness Levels")
plt.ylabel("Number of Pixels")

# Show all the figures and subplots
plt.show()
```
